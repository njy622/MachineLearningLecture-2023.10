## 로지스틱 회귀(Logistic Regression) - 분류

```python
import numpy as np
import matplotlib.pyplot as plt
```

- Sigmoid 함수

```python
def sigmoid(x, a=1, b=0):
    return 1. / (1+ np.exp(-a * (x - b)))
```

```python
xs = np.linspace(-5, 5, 1001)       #-5~5까지 1001개의 데이터
ys = sigmoid(xs)
plt.figure(figsize=(5,3))
plt.plot(xs, ys, label='sigmoid')
plt.plot(xs, ys*(1-ys), label='derivative')         #derivative: 미분
plt.yticks([0, 0.5, 1])
plt.title('Sigmoid function')
plt.grid(), plt.legend()
plt.show()
```

```python
# a 값에 따른 변화 (기울기의 정도, 가파른정도)

xs = np.linspace(-5, 5, 1001)     
y3 = sigmoid(xs, a=3)
y_half = sigmoid(xs, a=0.5)
plt.figure(figsize=(5,3))
plt.plot(xs, ys, label='a=1')
plt.plot(xs, y3, label='a=3')
plt.plot(xs, y_half, label='a=0.5')
plt.yticks([0, 0.5, 1])
plt.title('Sigmoid function')
plt.grid(), plt.legend()
plt.show()
```

```python
# b 값에 따른 변화

xs = np.linspace(-5, 5, 1001)     
y_plus2 = sigmoid(xs, b=2)
y_minus2 = sigmoid(xs, b=-2)
plt.figure(figsize=(5,3))
plt.plot(xs, ys, label='b=0')
plt.plot(xs, y_plus2, label='b=2')
plt.plot(xs, y_minus2, label='b=-2')
plt.yticks([0, 0.5, 1])
plt.title('Sigmoid function')
plt.grid(), plt.legend()
plt.show()
```
